```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
file_path = "C:/Users/ettor/OneDrive/Documenti/UNITN - TRIENNALE/CORSI/3 anno, 1 SEM/Strategia d'impresa/Gruppo 11 - Report/Analisi dati/input/dati.csv"
file = pd.read_csv(file_path)
```

```python
dataset = file 
dataset.head()
```

```python
dataset = dataset.rename(columns = {dataset.columns[1] : dataset.columns[1].strip()})
```

```python
dataset['Anno'] = dataset['Anno'].astype(str)
dataset = dataset.apply(lambda x: x.str.replace(",", ".") if isinstance(x, object) else x)
dataset['OUTPUT in trillion USD'] = dataset['OUTPUT in trillion USD'].astype(float)
dataset['OUTPUT GROWTH in percent'] = dataset['OUTPUT GROWTH in percent'].astype(float)
dataset['NUMBER OF ENTERPRISES in thousand'] = dataset['NUMBER OF ENTERPRISES in thousand'].astype(float)
```

```python
dataset.info()
```

```python
plt.figure(figsize=(10,8))

# Plot delle barre
bars = plt.bar(dataset['Anno'], dataset['OUTPUT in trillion USD'], color='steelblue', zorder = 2)

# Aggiunta dei valori sopra le barre
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom', ha='center')

# Settiamo il Titolo e il sottotitolo: 
plt.title('Worldwide (trillion USD ($))', fontsize=13)
plt.suptitle('Electronic Products & Components - Output', fontsize=12)  # y imposta la posizione verticale
plt.ylabel('In trillion USD ($)')

# Settiamo la griglia 
plt.grid(axis = 'y', alpha= 0.3, zorder = 1)
plt.xlim(-0.7, len(dataset['Anno']) - 0.2)

# nota 
plt.annotate('Grafico realizzato appositamente per il report con matplotlib.pyplot', xy=(0.80, -0.1), xycoords='axes fraction',
             fontsize=8, ha='center', va='center', color='black')
# nota 2 
plt.annotate('Source: Statista Market Insights', xy=(0.1, -0.1), xycoords='axes fraction',
             fontsize=10, ha='center', va='center', color='gray')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))

# Plot delle barre
plt.scatter(dataset['Anno'], dataset['OUTPUT GROWTH in percent'], color='steelblue', zorder = 2)
plt.plot(dataset['Anno'], dataset['OUTPUT GROWTH in percent'], color='steelblue', zorder = 2)

# height set:
for i, (x,y) in enumerate(zip(dataset['Anno'], dataset['OUTPUT GROWTH in percent'])):
    plt.text(x, y, f'{y}', ha = 'right', va = 'bottom', fontsize = 10, color = 'black', zorder = 3)

# Settiamo il Titolo e il sottotitolo: 
plt.title('Worldwide (percent)', fontsize=13)
plt.suptitle('Electronic Products & Components - Output Growth', fontsize=12)  # y imposta la posizione verticale
plt.ylabel('In percent')

# Settiamo la griglia 
plt.grid(axis = 'y', alpha= 0.3, zorder = 1)

# nota 
plt.annotate('Grafico realizzato appositamente per il report con matplotlib.pyplot', xy=(0.80, -0.1), xycoords='axes fraction',
             fontsize=8, ha='center', va='center', color='black')
# nota 2 
plt.annotate('Source: Statista Market Insights', xy=(0.1, -0.1), xycoords='axes fraction',
             fontsize=10, ha='center', va='center', color='gray')

plt.tight_layout()
plt.show()
```
